# for loop

```python
# for loop allow you to iterate through a sequence
```

```python
seq = [1,2,3,4,5]
```

```python
for i in seq:
    outcome = 'hello {}'.format(i)
    print(i)
    print(outcome)
    print('hello {}'.format(i))
```

# while loop

```python
# while loop used to continually perform an action untill some condition has been met
```

```python
i = 1;
while i<5:
    print('i is {}'.format(i))
```

```python
i = 1;
while i<=5:
    print('i is {}'.format(i))
    i = i+1
```

# list comprehension

```python
x = [1,2,3,4]
```

```python
null = []

for j in x:
    null.append(j**2)
```

```python
print(null)
```

```python
# below is the list comprehension, to breakdown the for loop into one simple line
```

```python
[5+j for j in x]
```

```python
out = [j**2 for j in x]
print(out)
```

# functions

```python
# function allow you to basically not to continually write code over and over again
```

```python
# you can just write it inside of a function and then call that function itself
```

```python
# the keyword of a function is def to define a function
```

```python
def func(parameter):
    print(parameter)
```

```python
func('hello')
```

```python
def func2(name):
    print('hi '+name)
```

```python
func2('fahri')
```

```python
func2('yazid')
```

```python
def func3(name='Apis',age=22): # if i give value here, it'll become a default value
                               # You can still replace the value when calling the function
    print('my name is '+name+' and i am {} years old'.format(age))
```

```python
func3()
```

```python
func3('fahri',23)
```
